```python
import gensim
from gensim.models import word2vec
import logging
import iterator_and_text_cleaner as it
```

```python
output_folder = '../converted/'
vector_size = 200
window_size = 5
words_per_article = 300
```

```python
# sentences_iterator = it.NewsIterator(output_folder, max_words=words_per_article, padding_size=window_size)
sentences_iterator = it.SimpleIterator(output_folder)
```

```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
# sg = 1 -> skipgram. sg=0 -> CBOW
# size -> Dimención de features a la salida (Palabras)
# min_count -> si no aparece mas de min_count, la palabra no forma parte del diccionario
# iter -> Cantidad de epochs
model = gensim.models.Word2Vec(sentences_iterator, iter=100, min_count=10, size=vector_size, workers=4, sg=1)
```

```python
model.save('word2vec.model')
```
